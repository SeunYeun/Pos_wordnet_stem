```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_purity_entropy_from_file(csv_path):
    df = pd.read_csv(csv_path)

    variants = df["variant"].unique()
    k_values = sorted(df["k"].unique())

    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    axes = axes.flatten()

    # Plot 1: KMeans - Purity
    for variant in variants:
        subset = df[(df["method"] == "KMeans") & (df["variant"] == variant)]
        axes[0].plot(subset["k"], subset["purity"], label=variant, marker="o")
    axes[0].set_title("KMeans - Purity")
    axes[0].set_xlabel("k")
    axes[0].set_ylabel("Purity")
    axes[0].legend()
    axes[0].grid(True)

    # Plot 2: KMeans - Entropy
    for variant in variants:
        subset = df[(df["method"] == "KMeans") & (df["variant"] == variant)]
        axes[1].plot(subset["k"], subset["entropy"], label=variant, marker="o")
    axes[1].set_title("KMeans - Entropy")
    axes[1].set_xlabel("k")
    axes[1].set_ylabel("Entropy")
    axes[1].legend()
    axes[1].grid(True)

    # Plot 3: BisectingKMeans - Purity
    for variant in variants:
        subset = df[(df["method"] == "BisectingKMeans") & (df["variant"] == variant)]
        axes[2].plot(subset["k"], subset["purity"], label=variant, marker="o")
    axes[2].set_title("BisectingKMeans - Purity")
    axes[2].set_xlabel("k")
    axes[2].set_ylabel("Purity")
    axes[2].legend()
    axes[2].grid(True)

    # Plot 4: BisectingKMeans - Entropy
    for variant in variants:
        subset = df[(df["method"] == "BisectingKMeans") & (df["variant"] == variant)]
        axes[3].plot(subset["k"], subset["entropy"], label=variant, marker="o")
    axes[3].set_title("BisectingKMeans - Entropy")
    axes[3].set_xlabel("k")
    axes[3].set_ylabel("Entropy")
    axes[3].legend()
    axes[3].grid(True)

    plt.tight_layout()
    plt.show()

csv_path = "../Data/csv/results.csv"
plot_purity_entropy_from_file(csv_path)
```
